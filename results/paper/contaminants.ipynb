# Virus

```python
df = pd.read_csv('virus_abundance_data.csv', index_col=0)
df.fillna(0., inplace=True)
df.head()
```

```python
df_sig = df.drop('batch', axis=1).loc[:, df.drop('batch', axis=1).ge(10).any()]
df_binary = np.where(df_sig > 0, 0, 1)
df_binary = pd.DataFrame(df_binary, index=df_sig.index, columns=df_sig.columns)
df_binary['batch'] = df['batch']
df_binary.head()
```

## Chi-squared test for association of read count vs batch

```python
batch_counts = df_binary['batch'].value_counts()
```

```python
df_counts = df_binary.groupby('batch').sum()
df_counts.head()
```

```python
virus = 'Human herpesvirus 6B'
table = pd.concat([df_counts[virus], batch_counts - df_counts[virus],
                       batch_counts], axis=1)
table.rename(columns={virus: 'virus', 0: 'no virus', 'batch': 'row_total'}, inplace=True)
table.loc['col_total']= table.sum(numeric_only=True, axis=0)
table[table['no virus'] > 0]
```

```python
virus = 'Human herpesvirus 6B'
chi2, p, dof, ex = chi2_contingency(table.drop('row_total', axis=1).drop('col_total', axis=0), correction=False)
print(f'Test statistic: {chi2}')
print(f'Degree of freedom: {dof}')
print(f'P value: {p}')
```

```python
n = table.at["col_total", "row_total"]  # grand total 
exp_counts = table.astype(np.float) 
for index in exp_counts.index[:-1]:
    for col in exp_counts.columns[:-1]:
        exp_counts.at[index, col] = (table.at[index, 'row_total']) * (table.at['col_total', col]) / n
exp_counts = exp_counts.drop('row_total', axis=1).drop('col_total', axis=0)
exp_counts.head()
```

```python
table = table.drop('row_total', axis=1).drop('col_total', axis=0)
tstat = np.sum(((table - exp_counts) ** 2 / exp_counts).values) # chi-squared test statistic
dof = (len(table.columns) - 1) * (len(table.index) - 1) # degrees of freedom 
pval = 1 - chi2.cdf(tstat, dof)
print(f'Test statistic: {tstat}')
print(f'Degree of freedom: {dof}')
print(f'P value: {pval}')
```

```python
batch_pval_dict = {}
for virus in df_counts.columns:
    # make contingency table
    table = pd.concat([df_counts[virus], batch_counts - df_counts[virus],
                       batch_counts], axis=1)
    table.rename(columns={virus: 'virus', 0: 'no virus', 'batch': 'row_total'}, inplace=True)
    table.loc['col_total']= table.sum(numeric_only=True, axis=0)

    # create dataframe with expected counts
    n = table.at["col_total", "row_total"]  # grand total 
    exp_counts = table.astype(np.float) 
    for index in exp_counts.index[:-1]:
        for col in exp_counts.columns[:-1]:
            exp_counts.at[index, col] = (table.at[index, 'row_total']) * (table.at['col_total', col]) / n
    exp_counts = exp_counts.drop('row_total', axis=1).drop('col_total', axis=0)
    
    # chi-squared test
    table = table.drop('row_total', axis=1).drop('col_total', axis=0)
    tstat = np.sum(((table - exp_counts) ** 2 / exp_counts).values) # chi-squared test statistic
    dof = (len(table.columns) - 1) * (len(table.index) - 1) # degrees of freedom 
    pval = 1 - chi2.cdf(tstat, dof)
    print(f'{virus}\t{pval}')
    batch_pval_dict[virus] = pval
```

## Chi-squared test for association of read count vs cell type

```python
type_col = [x.split(' - ')[1] for x in list(df_binary.index)]
df_binary_copy = df_binary.drop('batch', axis=1)
df_binary_copy['ctype'] = type_col
df_binary_copy.head()
```

```python
type_counts = df_binary_copy['ctype'].value_counts()
```

```python
df_counts = df_binary_copy.groupby('ctype').sum()
df_counts.head()
```

```python
ctype_pval_dict = {}
for virus in df_counts.columns:
    # make contingency table
    table = pd.concat([df_counts[virus], type_counts - df_counts[virus],
                       type_counts], axis=1)
    table.rename(columns={virus: 'virus', 0: 'no virus', 'ctype': 'row_total'}, inplace=True)
    table.loc['col_total']= table.sum(numeric_only=True, axis=0)

    # create dataframe with expected counts
    n = table.at["col_total", "row_total"]  # grand total 
    exp_counts = table.astype(np.float) 
    for index in exp_counts.index[:-1]:
        for col in exp_counts.columns[:-1]:
            exp_counts.at[index, col] = (table.at[index, 'row_total']) * (table.at['col_total', col]) / n
    exp_counts = exp_counts.drop('row_total', axis=1).drop('col_total', axis=0)
    
    # chi-squared test
    table = table.drop('row_total', axis=1).drop('col_total', axis=0)
    tstat = np.sum(((table - exp_counts) ** 2 / exp_counts).values) # chi-squared test statistic
    dof = (len(table.columns) - 1) * (len(table.index) - 1) # degrees of freedom 
    pval = 1 - chi2.cdf(tstat, dof)
    print(f'{virus}\t{pval}')
    ctype_pval_dict[virus] = pval
```

```python
sig = 0.05 / len(batch_pval_dict)
for virus in batch_pval_dict:
    if batch_pval_dict[virus] < sig or ctype_pval_dict[virus] < sig:
        print(f'{virus}\t{batch_pval_dict[virus]}\t{ctype_pval_dict[virus]}')
```

```python
contam_count = 0
for virus in batch_pval_dict:
    if batch_pval_dict[virus] < sig or ctype_pval_dict[virus] < sig:
        contam_count += 1
print(f'# of contaminants: {contam_count}')
print(f'# of non-contaminants: {len(batch_pval_dict) - contam_count}')
```
